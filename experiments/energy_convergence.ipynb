```python
from IMP.npctransport import *
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_max_pb(folder_path):
    files = os.listdir(folder_path)
    max_pb = np.max([int(f[:-3]) for f in files if ("rmf" not in f) and ("hdf5" not in f) and ("orig" not in f)])
    pb_path = f"{folder_path}/{max_pb}.pb"
    return pb_path

def load_time_energy(pb_path):
    with open(pb_path, "rb") as f:
        output = Output()
        fstring = f.read()
        output.ParseFromString(fstring)
        time_ns = []
        energy = []
        for i in range(len(output.statistics.global_order_params)):
            time_ns.append(output.statistics.global_order_params[i].time_ns)
            energy.append(output.statistics.global_order_params[i].energy)
    return time_ns, energy


# def plot_energy(pb_paths, labels, title):
#     for i, pb in enumerate(pb_paths):
#         time, energy = load_time_energy(pb)
#         plt.plot(time, energy, label = labels[i])
#     plt.title(title)
#     plt.xlabel("Time (ns)")
#     plt.ylabel("Energy (kcal/mol)")
#     plt.legend()
#     plt.show()

def plot_energy(pb_paths, labels, title, colors=None, legend_order=None):
    lines = []
    legend_labels = []
    
    # If no colors provided, let matplotlib handle default colors
    if colors is None:
        colors = [None] * len(pb_paths)
    
    for i, (pb, color) in enumerate(zip(pb_paths, colors)):
        time, energy = load_time_energy(pb)
        line, = plt.plot(time, energy, label=labels[i], color=color)
        lines.append(line)
        legend_labels.append(labels[i])
        
    plt.title(title)
    plt.xlabel("Time (ns)")
    plt.ylabel("Energy (kcal/mol)")
    
    if legend_order is not None:
        ordered_lines = [lines[i] for i in legend_order]
        ordered_labels = [legend_labels[i] for i in legend_order]
        plt.legend(ordered_lines, ordered_labels)
    else:
        plt.legend()
    
    plt.show()
```

```python
data_folder_path = "/cs/usr/roi.eliasian/LabFolder/NpcTransportExperiment/HS-AFM-Dataset/dataset/full_0uM_1ns/2"
files = os.listdir(data_folder_path)
max_pb = np.max([int(f[:-3]) for f in files if ("rmf" not in f) and ("hdf5" not in f)])
print(max_pb)
pb_path = f"{data_folder_path}/{max_pb}.pb"
plot_energy(pb_path, "Energy over time, full NPC model, 0 m NTRs")
```
